## All Dataset : 'test13_all' -> Model 첫 시도 
#### Epoch 9/100 36/36 [==============================] - 22s 606ms/step - loss: 0.0038 - mae: 0.0430 - mse: 0.0038 - val_loss: 0.0037 - val_mae: 0.0476 - val_mse: 0.0037

```python
model_version = 'model_20220111_all_1'

with h5py.File('./Model/Model_test13_all/dataset_hsv_test13_all.hdf5', 'r') as hf:
    image_train = np.array(hf['x_train'][:, :, :, :])
    action_train = np.array(hf['y_train'][:, :])
    image_val = np.array(hf['x_val'][:, :, :, :])
    action_val = np.array(hf['y_val'][:, :])

# print(image_train)
# print(action_train)
# print(image_val)
# print(action_val)
    
(x_train, y_train), (x_test, y_test) = (image_train, action_train), (image_val, action_val)

x_train /= 255.
x_test /= 255.

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(48, 64, 3)),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(2)
])

optimizer = tf.keras.optimizers.RMSprop(0.001)

callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
model.compile(loss='mse', optimizer=optimizer, metrics=['mae', 'mse'])

model.summary()
history = model.fit(x=x_train, y=y_train, batch_size=256, epochs=100, validation_data=(x_test, y_test), callbacks=[callback])

print("\n")
print(model.evaluate(x_test, y_test))


# model.save('./script/Model/'+model_version)
tf.saved_model.save(model, './Model/Model_test13_all/'+model_version)

del model

# model = tf.keras.models.load_model('./script/Model/'+model_version)
# print(model.evaluate(x_test, y_test))
# model.summary()

model = tf.saved_model.load('./Model/Model_test13_all/'+model_version)
outputs = model(x_test[0:10])

print(outputs)
```
