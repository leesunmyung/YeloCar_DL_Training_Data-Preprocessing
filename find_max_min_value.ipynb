# steeringValue, throttleValue 최댓값, 최솟값 찾기 (-> 임계값 reference)

```python
steeringValue = 0
throttleValue = 0
max_steeringValue, min_steeringValue = 0, 0
max_throttleValue, min_throttleValue = 0, 0
```

```python
flag = 0     # steeringValue, throttleValue 초기화

for date in os.listdir() :     
    os.chdir('./' + date)
    print(os.getcwd())
    
    for driver in os.listdir() :
        os.chdir('./' + driver)
        print(os.getcwd())
        
        for file in os.listdir() :
            if file.endswith(".log"):
                fin = open(file, "r")
                lines = fin.readlines()   
                
                for line in lines[1:] :    # 맨 윗줄 제외(date time steeringValue throttleValue)
                    if not line:
                        break
                    
                    steeringValue = int(line.split(' ')[2])
                    throttleValue = int(line.split(' ')[3])
                    
                    if flag == 0 :
                        max_steeringValue, min_steeringValue = steeringValue, steeringValue
                        max_throttleValue, min_throttleValue = throttleValue, throttleValue
                        print("--------- Initialization of max / min steeringValue, max / min throttleValue ----------\n")
                        print("max_steeringValue : {}, min_steeringValue : {}".format(max_steeringValue, min_steeringValue))
                        print("max_throttleValue : {}, min_throttleValue : {}\n".format(max_throttleValue, min_throttleValue))
                        print("---------------------------------------------------------------------------------------\n")
                        flag = 1
                    
                    if steeringValue > max_steeringValue :
                        print("max_steeringValue changing : {} -> {} in file {}".format(max_steeringValue, steeringValue, file))
                        max_steeringValue = steeringValue
                    elif steeringValue == 0 :
                        print("steeringValue : {} in file {}".format(steeringValue, file))       
                    elif steeringValue < min_steeringValue :
                        print("min_steeringValue changing : {} -> {} in file {}".format(min_steeringValue, steeringValue, file))
                        min_steeringValue = steeringValue
                   
                    if throttleValue > max_throttleValue :
                        print("max_throttleValue changing : {} -> {} in file {}".format(max_throttleValue, throttleValue, file))
                        max_throttleValue = throttleValue
                    elif throttleValue == 0 :
                        print("throttleValue : {} in file {}".format(throttleValue, file))         
                    elif throttleValue < min_throttleValue :
                        print("min_throttleValue changing : {} -> {} in file {}".format(min_throttleValue, throttleValue, file))
                        min_throttleValue = throttleValue
                        
                fin.close()
                
        os.chdir("../")
        
    os.chdir("../")

print("---------- Final max / min steeringValue, max / min throttleValue -----------\n")
print("max_steeringValue : {}, min_steeringValue : {}".format(max_steeringValue, min_steeringValue))
print("max_throttleValue : {}, min_throttleValue : {}\n".format(max_throttleValue, min_throttleValue))
print("-----------------------------------------------------------------------------\n")
```
