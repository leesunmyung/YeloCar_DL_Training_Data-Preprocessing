# 211222 폴더 포맷 맞춤

```python
for date in os.listdir() :  
    
        
    os.chdir('./' + date)
    print(os.getcwd())
    
    for driver in os.listdir() :
        os.chdir('./' + driver)
        print(os.getcwd())
        
        for file in os.listdir():
            if file.endswith(".log"):
                print(file)
                fin = open(file, "r")
                lines = fin.readlines()
                print(lines[0])
                
                if lines[0].startswith('date') :
                    continue
                    
                fout = open(file, "wt")
                fout.write('date time steeringValue throttleValue \n')
                
                for line in lines :
                    print(line)
                    if not line:
                        break
                    if line.startswith('date') :
                        continue
                    new_line = line.split('/')[0].split(' ')[0] + ' ' + line.split('/')[0].split(' ')[1] + ' ' + line.split('/')[0].split(' ')[4] + ' ' + line.split('/')[1].split(' ')[3]
                    print(new_line)
                    print(line)
                    fout.write(new_line + '\n')
      
                fin.close()
                fout.close()
                
        os.chdir("../")
        
    os.chdir("../")
```
